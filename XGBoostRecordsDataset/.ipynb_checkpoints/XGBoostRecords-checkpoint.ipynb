```python
import json
import glob
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
import shap
import tqdm
from IPython.display import display
```

```python
# ============================================================
# 1. LOAD GAMES
# ============================================================

def load_json_files(path_pattern):
    files = glob.glob(path_pattern)
    data = []
    for f in tqdm.tqdm(files):
        try:
            with open(f, "r", encoding="utf-8") as f_in:
                data.append(json.load(f_in))
        except UnicodeDecodeError as e:
            print(f"⚠️ Encoding error in file {f}: {e}. Skipping this file.")
        except json.JSONDecodeError as e:
            print(f"⚠️ JSON decode error in file {f}: {e}. Skipping this file.")
    return data


all_games = load_json_files("../aoe2dataset-main/rec_games_all_in_one_folder/*.recinfo")

# ============================================================
# 2. FEATURE ENGINEERING
#    - keep important building types
#    - keep age
#    - keep temporal deltas
# ============================================================

IMPORTANT_BUILDING_KEYS = [
    "townCenterCount",
    "castleCount",
    "barracksCount",
    "archeryRangeCount",
    "stableCount",
    "siegeWorkshopCount",
    "monasteryCount",
    "universityCount",
    "dockCount",
    "houseCount",
    "watchTowerCount",
    "guardTowerCount",
    "keepCount",
    "stoneWallCount",
    "palisadeWallCount",
]

def summarize_units(unitSummary):
    """Sum all unit counts into one total military number."""
    return sum(unitSummary.values())

def summarize_buildings(buildingSummary):
    """Total building count."""
    return sum(buildingSummary.values())

def extract_player_features(player):
    """Extract raw per-player features, with detailed buildings and age."""
    u = player["gameObjectSummary"]["unitSummary"]
    b = player["gameObjectSummary"]["buildingSummary"]
    r = player["resources"]

    feats = {}

    # --- Aggregate military ---
    feats["military_total"] = summarize_units(u)

    # --- Aggregate buildings ---
    feats["building_total"] = summarize_buildings(b)

    # --- Important building types as separate features ---
    for bk in IMPORTANT_BUILDING_KEYS:
        feats[f"b_{bk}"] = b.get(bk, 0)

    # --- Resources & eco/military summary ---
    feats["food"] = r["food"]
    feats["wood"] = r["wood"]
    feats["gold"] = r["gold"]
    feats["stone"] = r["stone"]

    feats["villagerCount"] = r["villagerCount"]
    feats["militaryCount"] = r["militaryCount"]

    feats["population"] = r["population"]
    feats["populationUntilHoused"] = r["populationUntilHoused"]

    # --- Tech + performance stats ---
    feats["age"] = r["age"]               # Dark/Feudal/Castle/Imperial
    feats["kills"] = r["kills"]
    feats["deaths"] = r["deaths"]
    feats["buildingsRazed"] = r["buildingsRazed"]
    feats["totalResearchCount"] = r["totalResearchCount"]
    feats["mapExplored"] = r["mapExplored"]

    feats["totalFood"] = r["totalFood"]
    feats["totalWood"] = r["totalWood"]
    feats["totalGold"] = r["totalGold"]
    feats["totalStone"] = r["totalStone"]

    feats["totalCastles"] = r["totalCastles"]
    feats["relicCount"] = r["relicCount"]
    feats["relicGold"] = r["relicGold"]

    return feats

def build_feature_rows_from_game(game_json, game_index, momentum_window=3):
    """
    For each game, build rows with:
      - raw p1 / p2 features (two real players, Gaia ignored)
      - current differences (p1 - p2)
      - temporal rates for p1 and p2 (1-step)
      - momentum over last `momentum_window` snapshots for p1 and p2
      - winner label: 1 if p1_name == winner_name else 0
    """
    rows = []

    winner_name = game_json.get("winner", None)

    # --- Figure out which two names are the actual players (ignore Gaia) ---
    p1_name = None
    p2_name = None

    for state in game_json["gameStates"]:
        players = state["playerMain"]["players"]
        real_players = [p for p in players if p.get("name") != "Gaia"]
        if len(real_players) >= 2:
            p1_name = real_players[0]["name"]
            p2_name = real_players[1]["name"]
            break

    # If we couldn't find two real players, skip this game
    if p1_name is None or p2_name is None:
        return rows  # empty list

    # Winner label: does p1_name match winner_name?
    winner_flag = 0 if (winner_name is not None and winner_name == p1_name) else 1

    # Histories for momentum
    history_p1 = []
    history_p2 = []
    history_time = []

    for state in game_json["gameStates"]:
        t = state["time"]
        players = state["playerMain"]["players"]

        # Build a quick lookup by name for this snapshot
        name_to_player = {p["name"]: p for p in players}

        # Ensure both players exist in this snapshot
        if p1_name not in name_to_player or p2_name not in name_to_player:
            # In theory shouldn't happen in normal recs, but be safe
            continue

        p1_player = name_to_player[p1_name]
        p2_player = name_to_player[p2_name]

        p1 = extract_player_features(p1_player)
        p2 = extract_player_features(p2_player)

        # ---- Current diff (snapshot-level advantage) ----
        diff = {f"{k}_diff": p1[k] - p2[k] for k in p1.keys()}

        # ---- Temporal: 1-step rates for p1 and p2 ----
        if len(history_p1) == 0:
            p1_delta = {f"p1_{k}_delta": 0.0 for k in p1.keys()}
            p2_delta = {f"p2_{k}_delta": 0.0 for k in p2.keys()}
        else:
            prev_p1 = history_p1[-1]
            prev_p2 = history_p2[-1]
            prev_t  = history_time[-1]
            dt = t - prev_t
            if dt == 0:
                dt = 1.0  # avoid /0 if timestamps equal

            p1_delta = {
                f"p1_{k}_delta": (p1[k] - prev_p1[k]) / dt
                for k in p1.keys()
            }
            p2_delta = {
                f"p2_{k}_delta": (p2[k] - prev_p2[k]) / dt
                for k in p2.keys()
            }

        # ---- Temporal: momentum over last N snapshots for p1 and p2 ----
        if len(history_p1) < momentum_window:
            p1_mom = {f"p1_{k}_mom": 0.0 for k in p1.keys()}
            p2_mom = {f"p2_{k}_mom": 0.0 for k in p2.keys()}
        else:
            old_p1 = history_p1[-momentum_window]
            old_p2 = history_p2[-momentum_window]
            old_t  = history_time[-momentum_window]
            dt_win = t - old_t
            if dt_win == 0:
                dt_win = 1.0

            p1_mom = {
                f"p1_{k}_mom": (p1[k] - old_p1[k]) / dt_win
                for k in p1.keys()
            }
            p2_mom = {
                f"p2_{k}_mom": (p2[k] - old_p2[k]) / dt_win
                for k in p2.keys()
            }

        # ---- Final row ----
        row = {
            "game_id": game_index,
            "time": t,
            "winner": winner_flag,   # 0 if p1_name == winner_name else 1
            # raw p1 / p2 state
            **{f"p1_{k}": v for k, v in p1.items()},
            **{f"p2_{k}": v for k, v in p2.items()},
            # current relative advantage
            **diff,
            # temporal info
            **p1_delta,
            **p2_delta,
            **p1_mom,
            **p2_mom,
        }

        rows.append(row)

        # update histories
        history_p1.append(p1)
        history_p2.append(p2)
        history_time.append(t)

    return rows

```

```python
# Build full dataframe
dataset_rows = []
for gi, g in enumerate(all_games):
    dataset_rows.extend(build_feature_rows_from_game(g, gi))

df = pd.DataFrame(dataset_rows)
print("Dataset shape:", df.shape)

# ============================================================
# 3. TRAIN/TEST SPLIT BY GAME
# ============================================================

unique_games = df["game_id"].unique()
train_games, test_games = train_test_split(unique_games, test_size=0.2, random_state=42)

train_df = df[df["game_id"].isin(train_games)]
test_df  = df[df["game_id"].isin(test_games)]

X_train = train_df.drop(columns=["winner", "game_id"])
y_train = train_df["winner"]

X_test = test_df.drop(columns=["winner", "game_id"])
y_test = test_df["winner"]

print("Train shape:", X_train.shape, "Test shape:", X_test.shape)
```

```python
pd.set_option('display.max_columns', None) # Show all columns
display(X_train)
```

```python
y_train.value_counts()
```

```python
# ============================================================
# 4. XGBOOST MODEL
# ============================================================

model = XGBClassifier(
    n_estimators=300,
    learning_rate=0.1,
    max_depth=6,
    subsample=0.8,
    colsample_bytree=0.8,
    eval_metric="logloss",
    n_jobs=-1,
    tree_method="hist"
)

model.fit(
    X_train,
    y_train,
    eval_set=[(X_train, y_train)],  # monitor training loss
    verbose=10                      # print progress every 10 trees
)

# ============================================================
# 5. EVALUATION
# ============================================================

# hard predictions
y_pred = model.predict(X_test)

# probabilities (P(p2 wins))
y_proba = model.predict_proba(X_test)[:, 1]

acc = accuracy_score(y_test, y_pred)
print("Accuracy:", acc)

from sklearn.metrics import roc_auc_score
auc = roc_auc_score(y_test, y_proba)
print("ROC AUC:", auc)

# ============================================================
# 6. FEATURE IMPORTANCE (SHAP)
# ============================================================

explainer = shap.TreeExplainer(model)

# subset for speed if many rows
sample_X = X_train.sample(min(2000, len(X_train)), random_state=0)
shap_values = explainer.shap_values(sample_X)

shap.summary_plot(shap_values, sample_X, max_display=30)

# ============================================================
# 7. SAVE MODEL
# ============================================================

model.save_model("aoe2_win_predictor.json")
print("Model saved!")
```

```python
import numpy as np

def collapse_time(game_df, threshold=0.9):
    probs = model.predict_proba(
        game_df.drop(columns=["winner", "game_id"])
    )[:, 0]
    
    idx = np.where(probs >= threshold)[0]
    if len(idx) == 0:
        return 1.0  # never confident until end
    first_idx = idx[0]
    return game_df["time"].iloc[first_idx] / game_df["time"].iloc[-1]  # normalized

results = []
for gid in df["game_id"].unique():
    game_df = df[df["game_id"] == gid].sort_values("time")
    results.append(collapse_time(game_df))

print("Avg certainty time:", np.mean(results))
print("Median certainty time:", np.median(results))
```

```python
plt.hist(results, bins=20)
plt.xlabel("Fraction of game until certainty")
plt.ylabel("Count")
plt.title("Distribution of Collapse/Certainty Times")
plt.grid(True)
plt.show()
```

```python
def point_of_no_return_backward(game_df, feature_cols, threshold=0.8):
    game_df = game_df.sort_values("time")
    n_rows = len(game_df)

    winner_label = int(game_df["winner"].iloc[-1])   # 0 = p1, 1 = p2

    X_game = game_df[feature_cols]
    p1_probs = model.predict_proba(X_game)[:, 0]     # P(p1 wins)

    if winner_label == 0:
        conf = p1_probs           # confidence P1 wins
    else:
        conf = 1.0 - p1_probs     # confidence P2 wins

    if conf[-1] < threshold:
        return 1.0

    i = n_rows - 1
    while i >= 0 and conf[i] >= threshold:
        i -= 1

    if i < 0:
        ponr_idx = 0
    else:
        ponr_idx = i + 1

    t_norm = game_df["time"].iloc[ponr_idx] / game_df["time"].iloc[-1]
    return float(t_norm)
```

```python



ponr_list = []
for gid in df["game_id"].unique():
    g = df[df["game_id"] == gid]
    ponr_list.append(point_of_no_return_backward(g, feature_cols, threshold=0.97))

print("Mean PONR:", np.mean(ponr_list))
print("Median PONR:", np.median(ponr_list))

import matplotlib.pyplot as plt
plt.hist(ponr_list, bins=20)
plt.xlabel("Fraction of game at point-of-no-return")
plt.ylabel("Number of games")
plt.title("Relaxed Backward Point-of-No-Return Distribution")
plt.grid(True)
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np

# pick a game
game_id = df["game_id"].unique()[10]

# get game data (we won't add new columns here)
game_df = df[df["game_id"] == game_id].sort_values("time")

# model input: ONLY the columns it was trained on
X_game = game_df[feature_cols]

# win probability for p1
p1_prob = model.predict_proba(X_game)[:, 0]

# features to overlay (must be in feature_cols)
features_to_plot = [
    "p1_villagerCount",
    "p1_totalGold",
    "p1_kills",
    "p1_population"
]

time_vals = game_df["time"].values

plt.figure(figsize=(14, 7))

# --- plot probability ---
plt.plot(time_vals, p1_prob, label="P(p1 wins)", color="black", linewidth=2)

# --- loop normalized features ---
for feat in features_to_plot:
    if feat not in game_df.columns:
        print(f"Skipping {feat}: not in dataframe")
        continue
    
    vals = game_df[feat].values.astype(float)
    vmin, vmax = vals.min(), vals.max()
    if vmax > vmin:
        vals_norm = (vals - vmin) / (vmax - vmin)
    else:
        vals_norm = np.zeros_like(vals)
    
    plt.plot(
        time_vals,
        vals_norm,
        linestyle="--",
        linewidth=1.8,
        label=f"{feat} (norm)"
    )

plt.axhline(0.5, color="gray", linestyle=":", alpha=0.6)

plt.xlabel("Time")
plt.ylabel("Normalized Feature Value / Probability")
plt.title(f"Win Probability and Feature Trajectories (Game {game_id})")
plt.legend(loc="upper left", bbox_to_anchor=(1.05, 1.0))
plt.grid(True)
plt.tight_layout()
plt.show()
```
